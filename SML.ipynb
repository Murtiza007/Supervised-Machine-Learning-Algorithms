```python
import pandas as pd
data=pd.read_csv('1000_Companies.csv')
data.head()
```

```python
#forest Tree Classification

import pandas as pd
data=pd.read_csv('Iris.csv')

data.shape
data.value_counts('Species')
data.head()
```

```python
x=data.iloc[:,1:5].values
y=data.iloc[:,5].values
x,y
```

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
y=lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=50)
classifier.fit(x_train,y_train)
```

```python
accuracy=classifier.score(x_test,y_test)
accuracy
```

```python
new_data=[4.9,	3.0	,1.4,	0.2	]
predict=classifier.predict([new_data])

if (predict[0]==[1,0,0]).all():
  print('setosa')
elif (predict==[0,1,0]).all():
  print('versicolor')
elif (predict==[0,0,1]).all():
  print('virginica')
else:
  print('unkown species')
```
